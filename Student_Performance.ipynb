```python
x = df.drop(['exam_score'],axis=1)
y = df['exam_score']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),[0])
],remainder='passthrough')

step2 = LinearRegression()
```

```python
pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pre = pipe.predict(x_test)
```

```python
result = round(r2_score(y_test,pre)*100,2)
```

```python
result
```

```python
step1 = ColumnTransformer([
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),[0])
],remainder='passthrough')

step2 = DecisionTreeRegressor()
```

```python
pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pre = pipe.predict(x_test)
```

```python
result = round(r2_score(y_test,y_pre)*100,2)
```

```python
result
```

```python
step1 = ColumnTransformer([
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),[0])
],remainder='passthrough')

step2 = RandomForestRegressor()
```

```python
pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
predication = pipe.predict(x_test)
```

```python
result = round(r2_score(y_test,predication)*100,2)
```
